```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
plt.style.use('ggplot')
```

```python
stats = pd.read_csv('C:/Users/user/Downloads/archive/Men T20I Team Match Results - 21st Century.csv')
```

```python
stats.set_index('Country',inplace= True)
```

```python
stats
```

```python
stats['Result'].value_counts().plot(kind='line',figsize=(12,8))
                                   
plt.title("LINE REPRESENTATION OF RESULTS OF MATCHES",size=22)
plt.ylabel('Number Of Matches',size=18)
plt.xlabel('Result',size=18)
plt.show()
                                
```

```python
stats
```

```python
stats['Ground'].value_counts().sort_values(ascending=False)
```

```python
stats['Ground'].value_counts().sort_values(ascending=False).head(15).plot(kind='bar',figsize=(8,9),color=('black','yellow'))
plt.title('Matches Played in  various Grounds(top 15)',size=20)
plt.xlabel('Ground Name',size=14)
plt.ylabel('no. of matches played',size=14)
plt.show()
```

```python
stats['Match Year'].value_counts().plot(kind='pie',figsize=(8,15),autopct='%.2f%%',
                               labels=None,
                            startangle=90,
                               pctdistance=.5,
                            shadow=True)
plt.title('Number Of Matches played per year in pie',size=20)
plt.legend(labels=stats['Match Year'],loc='upper right')
plt.axis('equal')
plt.show()
```

```python
stats['Match Period'].value_counts().plot(kind='area',color=('orange','red'),figsize=(10,10))
plt.title('Number Of Matches during 2005 to 2020 per month',size=24,color=('maroon'))
plt.ylabel('Matches Played',size=16)
plt.xlabel('Period',size=16)
plt.show()

```

```python
stats['Margin'].value_counts().head(15).sort_values(ascending=True).plot(kind='barh',figsize=(12,15),color=('purple','yellow'))
plt.title('margin by which teams won',size=22)
plt.xlabel('Matches',size=16)
plt.ylabel('Margin',size=16)
plt.show()
```

```python
stats.index.value_counts().head(10).plot(kind='line',stacked=False,figsize=(10,14),color='darkgreen')
                                         
plt.title('no.of matches played my countries(top 10)',size=24)
plt.ylabel('matches played',size=16)
plt.xlabel('teams',size=16)
plt.show()
```

```python
drops = ['Matches','Match Date']
stats.drop(drops , axis=1 , inplace=True)
```

```python
stats['Match'].value_counts().head(15).plot(kind='pie',labels=None,
                            startangle=90,
                               pctdistance=1.08,
                            shadow=True,figsize=(15,8),autopct='%.2f%%')

plt.title('Most number Of Matches by teams played against opponets(top 15)',size=20)
plt.legend(labels=stats['Match'],loc='upper right')
plt.axis('equal')
plt.ylabel('MATCHES',size=20)
plt.show()
```
